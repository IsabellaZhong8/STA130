By examining these patterns, we can see how a strong bias for social media affects loneliness and overall wellness. Moreover, we’ll analyze the correlations between high social media usage and reported levels of loneliness and mental health, which provides insights to how online interactions may influence personal well-being and in-person social abilities.

<b>Visualization<b>:

This provides us with a better idea as to how much online interactions may affect us mentally, and gives us insight as to whether there is a correlation between the two. 
    
3. <b>Bootstrapping/histogram<b> 

Bootstrapping a simulation of samples, where we can calculate the r value between CONNECTION_social_media_time_per_day, and the two wellness variables: LONELY_ucla_loneliness_scale_companionship and WELLNESS_self_rated_mental_health. Analyzing these histogram will help show the nature of their relationships, ie. analyzing modality, confidence intervals, variance, standard deviation, etc.

<b>Assumption<b>: 

- All observations are independent and NOT biased

- The likelihood of getting someone with low levels of loneliness, high mental health and low social media use 
- The likelihood of getting someone with high scores of loneliness, low scores of mental health and high social media use 

Using our bootstrapped data, we can calculate the p values for each test and make an inference to whether we should reject the null or not. Our analysation from the graphs can also further help us insinuate to a conclusion. 
    
<b>Hypothesis of result<b>: 
    
I believe there that for tests a) and b), the p value will be significantly less than the significance level. Having p-values comparably less than the significance level indicates that there is strong evidence against the null, thus showing that there is a strong negative correlation between excessive interactions online and poor mental health.

<b>Visualization<b>

In a histogram, we can separate the groups of people in bins based on their financial standing (BELIEFS_rank_financial_situation), and set the y-axis as the measure of their comfort in talking with strangers (CONNECTION_comfort_talking_with_strangers)

<b>Overall analysis<b>: 

<b>Assumptions<b>: 
    
- Relationship between BELIEFS_rank_financial_situation and CONNECTION_comfort_talking_with_strangers is linear 
- Residuals are normally distributed 
- All observations are independent of one another 

We can model the correlation between using linear regression where the dependent variable would be the financial ranking, and the independent variable can be social comfort or effectiveness of creating connections in the workplace (in which case we would perform linear regression twice, one for each independent variable). Thus, we can use the linear regression model: Y = B0 + B1*xi, and evaluate what the slope (B1) of the line is using the least squares method (which helps minimize the sum of squared residuals - aka epsilon). 

Afterwards, we can run a hypothesis test, where the null (H0) is that there is no linear correlation (b1 = 0) - this shows whether the slope that we just found is statistically significant or the result of random static noise. Setting the significance level as 0.05 (under the assumption that the null is true), if the p-value is significantly below the significance level, it indicates there is a strong correlation between the two variables. To further demonstrate the strength of the correlation, we can find the R2 value, where the closer it is to 1, the stronger the correlation.

Furthermore, we can identify whether it is a positive or negative correlation based on the slope value. Since the hypothesis test has shown how B1 is statistically significant, then if B1 > 0, then there is a positive correlation, whereas if B1 < 0, then there is negative correlation. I hypothesize that B1 will be positive, thus allowing us to make the inference that there indeed is a linear correlation between connecting with others/social comfort and a prosperous life

<b>Analysis overview<b>: 
    
In this analysis, we are accounting for many variables to help estimate if COVID-19 has caused a negative impact on our generation. DEMO_age will play in as an independent variable to compare the results of our generation Z to the previous generation’s ability in communicating (PSYCH_social_interactions_anxiety_scale_talking). Additionally, we will observe how dramatic of an effect the ability to socialize with others was in relation to COVID-19 by analyzing the relationship between intensity of loneliness (in comparison to pre-covid) and levels of social anxiety. 
    
<b>Analysis methods<b>: 
    
<b>Visualization<b> 
1. <b>Scatterplot<b>

<b>Overall analysis<b>: 

<b>Assumptions<b>: 
    
- Relationship between DEMO_age and PSYCH_social_interactions_anxiety_scale_talking levels is linear 
- Residuals are normally distributed 
- All observations are independent of one another 

In this analysis, we can use permutation testing. We can define our null hypothesis (H0) to be that COVID-19 has not significantly affected any generation’s social and loneliness levels. Our test statistic that we will focus on is the correlation coefficient between loneliness post-covid and social anxiety. Thus, we can calculate the observed test statistic using a hypothetical mean of 5, and permutate the data (following the same procedure of calculating the test statistic), which will create a null distribution. Then, we can find the p-value (which is the % of permutation test statistics that are as or more extreme than the observed test statistic). Under the assumption that the null is true, if the p-value is less than the significance value (0.05), then we can reject the null hypothesis. 
Moreover, our results from analyzing the scatterplot (linear regression) can demonstrate how great of an impact COVID-19 has put on Generation Z vs. previous generations. We can do this by forming the null-hypothesis that Generation Z and other generations have been affected similarly in terms of social skills by COVID-19. 
For scatter plots, we can follow the procedure of finding the slope (B1) and p-value under the assumption that the null is true. If p-value is less than significance level, then it shows that the slope is statistically significant. Also, if B1 is positive, then it shows that the levels of social anxiety decrease as age increases, while if it is negative, then it demonstrates the opposite. 


<b>Hypothesis of result<b>: 
    
I believe that we will find a p-value that is less than the significance value (0.05), which indicates that there is strong evidence against the null that: COVID-19 has not significantly affected any generation’s social and loneliness level. Thus, we can make the inference that COVID-19 has made an impact on our generation. This can also be applied to the scatterplot. Moreover, I think that R2 will be close to 1, showing that the correlation between age and social levels is strong. I also think that B1 will be positive, demonstrating the older someone is, the higher their social levels are. Thus, we can also make the inference that COVID-19 had a greater impact on Generation Z than previous generations in terms of social skills.
